```python
gdp_df = pd.read_csv("resources/actual GDP.csv", skiprows=4)
select_gdp_df = gdp_df[['Country Name','Country Code','Indicator Name','Indicator Code','1990','2019']]
dropped_gdp_df= select_gdp_df.dropna()
```

```python
# clean Population data
pop_df = pd.read_csv("resources/pop.csv", skiprows=4)
select_pop_df = pop_df[['Country Name','Country Code','Indicator Name','Indicator Code','1990','2019']]
select_pop_df = select_pop_df.dropna()
```

```python
clean_pop_df = select_pop_df.drop(select_pop_df[select_pop_df['2019'] < 500000].index)
clean_pop_df
```

```python
clean_pop_df['POP_change']=clean_pop_df['2019']-clean_pop_df['1990']
clean_pop_df
```

```python
population_df= pd.DataFrame(clean_pop_df)
population_df.to_csv("Resources/population.csv")
```

```python
Net_Migration =pd.read_csv("resources/Net Migration.csv", skiprows=4)
```

```python
Net_Migration.drop(['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
       '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '2020', '2021', 'Total Sum'], axis =1, inplace = True)
             
```

```python
country_names_df = Net_Migration[['Country Name','Country Code','Indicator Name','Indicator Code']]
country_names_df
```

```python
Net_Migration
```

```python
select_net_migration_df = Net_Migration.sum(axis = 1)
select_net_migration_df
```

```python
country_names_df['net_migration'] = select_net_migration_df
country_names_df
```

```python
Migration_df= pd.DataFrame(country_names_df)
Migration_df.to_csv("Resources/Net_mig_nat_bir.csv")
```

```python
nat_birth_rate =pd.read_csv("resources/Birth rate.csv", skiprows=4)
nat_birth_rate.drop(['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
       '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '2020', 'Average'], axis =1, inplace = True)
mean_nat_birth_rate = nat_birth_rate.mean(axis = 1)
mean_nat_birth_rate
```
